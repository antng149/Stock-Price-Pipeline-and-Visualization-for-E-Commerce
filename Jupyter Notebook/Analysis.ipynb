```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
project3_df = pd.read_csv('results.csv')
```

```python
project3_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization 1: Line plot of average volatility by date
plt.figure(figsize=(10, 6))
sns.lineplot(data=project3_df, x='date', y='average_volatility', hue='name')
plt.title('Average Volatility by Date')
plt.xlabel('Date')
plt.ylabel('Average Volatility')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Costco is the most volatile !!

```python
# Group the data by company name and date, and calculate the maximum value of highest volatility for each group
grouped_df = project3_df.groupby(['name', 'date'])['highest_volatility'].max().reset_index()

# Create a list of unique dates
dates = grouped_df['date'].unique()

# Set the size of the plot
plt.figure(figsize=(12, 6))

# Get the unique companies in the data
companies = grouped_df['name'].unique()

# Set the width of each bar
bar_width = 0.8 / len(dates)

# Iterate through each date and plot the highest volatility for each company as a bar
for i, date in enumerate(dates):
    date_data = grouped_df[grouped_df['date'] == date]
    # Calculate the x positions for each bar of the date
    x_pos = [j + i * bar_width for j in range(len(companies))]
    plt.bar(x_pos, date_data['highest_volatility'], width=bar_width, label=date, alpha=0.8)

# Set the title, x-axis label, y-axis label, and legend
plt.title('Daily Highest Volatility per Company', fontsize=16)
plt.xlabel('Company', fontsize=14)
plt.ylabel('Highest Volatility', fontsize=14)
plt.legend()

# Set the x-axis ticks and labels to the unique companies
plt.xticks([i + bar_width * len(dates) / 2 for i in range(len(companies))], companies, rotation=90)

# Show the plot
plt.show()

```

Yes, the finding of this graph support my previous conclusion from the first graph. Costco is the most volatile.
